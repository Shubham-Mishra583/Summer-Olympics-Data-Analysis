# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np

df = pd.read_csv('summer.csv')
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data = []

for SportMedal,info in df.groupby(["Sport","Medal"]):
     if SportMedal[1] == "Gold":
            data.append([SportMedal[0],len(info)])
data = pd.DataFrame(data,columns=['Sports','Gold Medals']).sort_values(by='Gold Medals',ascending=False).head()
data.reset_index(inplace=True,drop=True)
print(data)
data.plot(x = "Sports",y = "Gold Medals",kind = "bar" )
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []
for Sport,info in df.groupby('Sport'):
    data.append([Sport,len(info)])
data = pd.DataFrame(data,columns=['Sport','Total Medals']).sort_values(by='Total Medals',ascending=False).head()
data.reset_index(inplace=True,drop=True)
print(data)
data.plot(x = "Sport",y = "Total Medals",kind = "bar" )
```

### 4. Which player has won most number of medals? (Top 5)

```python
data = []
for Athlete,info in df.groupby('Athlete'):
    data.append([Athlete,len(info)])

data = pd.DataFrame(data,columns=['Athlete','Total Medals']).sort_values(by='Total Medals',ascending=False).head()
data.reset_index(inplace=True,drop=True)
print(data)
data.plot(x = "Athlete",y = "Total Medals",kind = "bar" )
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
data = []

for AthleteMedal,info in df.groupby(['Athlete','Medal']):
    if AthleteMedal[1] == "Gold":
        data.append([AthleteMedal[0],len(info)])

data = pd.DataFrame(data,columns=['Athlete','Gold Medals']).sort_values(by='Gold Medals',ascending=False).head()
data.reset_index(inplace=True,drop=True)
print(data)
data.plot(x = "Athlete",y = "Gold Medals",kind = "bar") 
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []

for Event in df["Event"].unique():
    data.append([Event,len(df[df["Event"]==Event])])
data = pd.DataFrame(data,columns=['Event','Athlete']).sort_values(by='Athlete',ascending=False).head()
data.reset_index(inplace=True,drop=True)
print(data)
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data = []

for SportGenderMedal,info in df.groupby(['Sport','Gender','Medal']):
    if SportGenderMedal[1] == 'Women' and SportGenderMedal[2] == 'Gold':
        data.append([SportGenderMedal[0],len(info)])
data= pd.DataFrame(data,columns=['Sport','Gold Medals']).sort_values(by='Gold Medals',ascending=False).head()
data.reset_index(inplace=True,drop=True)
print(data)
data.plot(x = "Sport",y = "Gold Medals",kind = "bar") 
```
